## Data Cleaning

```python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Replace common special characters (& and %) with their word equivalents
    text = text.replace('&', 'and').replace('%', 'percent')
    # Remove any character that is not a letter, number, or space
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove any extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df['Invoice Column'] = df['Invoice Column'].apply(clean_text)
df['Purchase Order Column'] = df['Purchase Order Column'].apply(clean_text)
```

```python
# Tokenization
def tokenize_text(text):
    # Tokenize the text by splitting on commas or spaces
    tokens = re.split(r',\s*|\s+', text)
    return tokens
```

```python
# Apply tokenization
df['Invoice Tokens'] = df['Invoice Column'].apply(tokenize_text)
df['Purchase Order Tokens'] = df['Purchase Order Column'].apply(tokenize_text)
```

```python
# Removing numbers and other unwanted tokens
def remove_noise(tokens):
    # Remove any token that is purely numeric
    tokens = [token for token in tokens if not token.isdigit()]
    return tokens
```

```python
df['Invoice Clean Tokens'] = df['Invoice Tokens'].apply(remove_noise)
df['Purchase Order Clean Tokens'] = df['Purchase Order Tokens'].apply(remove_noise)
```

```python
from sentence_transformers import SentenceTransformer, InputExample, losses, models
from torch.utils.data import DataLoader
from sentence_transformers import datasets
import random
```

```python
import ast
import random
from sentence_transformers import InputExample

def create_training_data(df):
    train_examples = []

    for idx, row in df.iterrows():
        # Convert the 'Mapped Result' from string to a dictionary (if necessary)
        mapped_result = row['Mapped Result']
        if isinstance(mapped_result, str):
            mapped_result = ast.literal_eval(mapped_result)  # Convert string to dictionary

        # Positive example: matching invoice-purchase order pair (from the "Mapped Result" column)
        for inv_item, po_item in mapped_result.items():  # Iterate through the dictionary
            train_examples.append(InputExample(texts=[inv_item, po_item], label=1.0))  # Positive example

        # Negative example: mismatched pair (randomly shuffle purchase orders and ensure they don't match)
        invoice_items = list(mapped_result.keys())  # Extract invoice items (keys from the dictionary)
        purchase_order_items = list(mapped_result.values())  # Extract purchase order items (values from the dictionary)

        # Shuffle the purchase orders to create mismatched pairs
        shuffled_po_items = random.sample(purchase_order_items, len(purchase_order_items))
        for inv_item, shuffled_po_item in zip(invoice_items, shuffled_po_items):
            # Ensure it's a mismatch
            if inv_item != shuffled_po_item:
                train_examples.append(InputExample(texts=[inv_item, shuffled_po_item], label=0.0))  # Negative example

    return train_examples
```

```python
train_examples = create_training_data(df)
```

```python
import pandas as pd

# Create a DataFrame from the train_examples
df_train_examples = pd.DataFrame({
    'Text1': [example.texts[0] for example in train_examples],
    'Text2': [example.texts[1] for example in train_examples],
    'Label': [example.label for example in train_examples]
})

# Print the DataFrame
df_train_examples.head(10)  # Print the first few rows
```

```python
# Load pre-trained Sentence-BERT model
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
# Prepare the DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
```

```python
# Define the loss function for fine-tuning
train_loss = losses.CosineSimilarityLoss(model)
```

```python
# Fine-tune the model
num_epochs = 3  # You can adjust this as per your needs
warmup_steps = int(len(train_dataloader) * num_epochs * 0.1)
```

```python
# Fine-tune the model
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=num_epochs,
    warmup_steps=warmup_steps,
    output_path='fine_tuned_invoice_purchase_sbert_model'  # Save model after training
)
```
